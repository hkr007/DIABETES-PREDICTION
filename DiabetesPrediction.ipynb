```python
from sklearn.preprocessing import StandardScaler

scalar=StandardScaler()

x_scaled=scalar.fit(x)
x_scaled=scalar.fit_transform(x)



#scalled the featurs 

#scaled bxz of some columns have 2 digits and some are 3 or 4 digits 
#so standard scaling to minimize the distance
```

```python
# in stats modules we have this 
#we can use co relation metrics also 
from statsmodels.stats.outliers_influence import variance_inflation_factor

```

```python
vif = pd.DataFrame()
vif["vif"] = [variance_inflation_factor(x_scaled,i)
              for i in range(x_scaled.shape[1])]
vif["Features"] = x.columns

#let's check the values
vif

#here vif<5 so no colinearity
```

```python
from sklearn.model_selection import train_test_split

from sklearn.linear_model import LogisticRegression

reg=LogisticRegression()

```

# implement different metics:

```python
from sklearn.metrics import confusion_matrix,accuracy_score,roc_curve,roc_auc_score



accuracy=accuracy_score(y_test,y_predict)
```

```python
accuracy
```

```python
confusion_mat=confusion_matrix(y_test,y_predict)
confusion_mat

#this is our confusion matric 

#here tp-112,fp-17,fn-26,tn-37
```

```python
tp=confusion_mat[0][0]
fp=confusion_mat[0][1]
fn=confusion_mat[1][0]
tn=confusion_mat[1][1]

#this bcz of we need to understand the all matrics 
```

```python
accuracy=(tp+tn)/(tp+tn+fp+fn)
accuracy

#getting same accuracy
```

```python
recall=tp/(tp+fn)
recall
```

```python
precsion=tp/(tp+fp)
precsion
```

```python
f2score=2+(precsion+recall)/precsion+recall

f2score

#this is mean here 
```

```python
auc=roc_auc_score(y_test,y_predict)
auc

#area under curve
#high auc means best model
```
